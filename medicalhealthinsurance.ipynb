**Split into Train and Test**

```python
X = insurance_df.drop('charges', axis = 1)   # Drop Y i.e output
X
```

```python
Y=insurance_df['charges']
Y
```

**Cross validation **

1.   Split data into train and test



```python
from sklearn.model_selection import train_test_split
#Random state between 0-9 role is to fix random sampling of data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=0)
X_train.shape
```

```python
X_test.shape
```

**ANN Regression**

```python

from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import RMSprop
```

```python
def build_model():
  model=Sequential()
  model.add(Dense(units = 128, activation='relu', input_shape = [len(X.keys())]))
  model.add(Dense(units=256,activation='relu'))
  model.add(Dense(units=512,activation='relu'))
  model.add(Dense(units=1))
  optimizers=RMSprop(learning_rate=0.001)
  model.compile(loss='mean_squared_error',optimizer=optimizers,metrices=['mean_squared_error', 'mean_absolute_error'])
  return model
```

```python
def build_model():
    # Sequential Neural Network
    model = Sequential()
    
    # Input Layer 
    # Dense - Dense Neural Network , activation function , input_shape = number of features
    # units = number of neurons
    model.add(Dense(units = 128, activation='relu', input_shape = [len(X.keys())]))

    # Hidden Layer - I
    model.add(Dense(units = 256, activation='relu'))

    # Hidden Layer - II
    model.add(Dense(units = 512, activation='relu'))

    # Output Layers - Regression = 1
    model.add(Dense(units = 1))
 # Optimizers : Reduce the loss
    optimizers = RMSprop (learning_rate = 0.001)

    # Model compiler
    model.compile(loss = 'mean_squared_error', optimizer = optimizers, metrics = ['mean_squared_error', 
                                                                                 'mean_absolute_error'])
    return model
```

```python
model = build_model()

model.summary()
```

```python
# epochs = Number of Iterations
# batch_size = number of samples per iteration
# Validation Split = 20 % of X_train & X_test : Testing of Model done simuntaneously per epoch

history = model.fit(X_train, Y_train, epochs = 600, batch_size= 30, validation_split = 0.2)
```

```python
pd.DataFrame(history.history)[['mean_squared_error','val_mean_squared_error']].plot()
```

**Predictions**

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score  #R^2 values denotes how much % value is getting matched
r2_score (Y_test, predictions)
```
